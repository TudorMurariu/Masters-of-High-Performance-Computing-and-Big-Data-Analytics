## Let's agradate every algorithm and see how they preforme on avg

```python
model_performance = df.groupby("id_alg")["mean_ind"].agg(["mean", "std", "max", "min"])
model_performance = model_performance.sort_values(by="mean", ascending=False)

model_performance
```

```python
model_performance = model_performance.sort_values(by="std", ascending=True)
print(model_performance)
```

It seeams taht algo 10 has the highest mean score while also having the lowest standard deviation, this indicates that it might be the best algo.

## Let's try to also visualise the data now

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(30, 10))

sns.lineplot(ax=axes[0], x=df.index, y=df["ind_0"])
axes[0].set_title("Index 0")

sns.lineplot(ax=axes[1], x=df.index, y=df["mean_ind"])
axes[1].set_title("Mean Index")

sns.lineplot(ax=axes[2], x=df.index, y=df["std_ind"])
axes[2].set_title("Standard Deviation Index")

fig.suptitle("Analysis of Model Performance", fontsize=30, fontweight='bold')
plt.show()
```

```python
sns.kdeplot(data=df, x='mean_ind', hue='id_alg', fill=True).set_title("Kernel Density Estimate")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="id_alg", y="mean_ind", data=df)
plt.xlabel("Model ID")
plt.ylabel("Mean Performance")
plt.title("Model Performance Distribution")
plt.xticks(rotation=45)
plt.show()
```

## Let's now go over each dataset and see how every algo does on them:


```python
algo_performance = df.groupby(["id_dataset", "id_alg"])["mean_ind"].agg(["mean", "std"])
best_algos = algo_performance.loc[algo_performance.groupby("id_dataset")["mean"].idxmax()]
print(best_algos)
```

```python
algo_performance = algo_performance.reset_index()
heatmap_data = algo_performance.pivot(index="id_dataset", columns="id_alg", values="mean")

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, cmap="viridis", linewidths=0.5)
plt.title("Algorithm Performance Across Datasets")
plt.xlabel("Algorithm ID")
plt.ylabel("Dataset ID")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for dataset_id in df["id_dataset"].unique():
    subset = algo_performance[algo_performance["id_dataset"] == dataset_id]
    plt.plot(subset["id_alg"], subset["mean"], marker="o", label=f"Dataset {dataset_id}")

plt.xlabel("Algorithm ID")
plt.ylabel("Mean Performance")
plt.title("Algorithm Performance Across Different Datasets")
plt.legend()
plt.show()
```

We can see in the plot above that algo 10 is the highest pick or one of the higest in most datasets.
